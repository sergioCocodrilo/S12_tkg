TKG Anlaysis
=============


Import the data

```python
import pandas as pd
df = pd.read_csv('tkg.csv')
```

```python
df.head()
```

List the trunk groups

```python
tkgs = df.tkgid.unique()
[print(i, tkg) for i, tkg in enumerate(tkgs)]
```

List all the possible states of the trunks
------

```python
all_trunk_states = df.state.unique()
print(all_trunk_states)
```

Get the count of each trunk state for each trunk group
-----

```python
for tkg in tkgs:
    for state in all_trunk_states:
        print(tkg, state, ':', len(df[(df.tkgid == tkg) & (df.state == state)].index))
    print()
```
